```python
#uefa league draw
class Club:
    def __init__(self, name: str, country: str, pot: str) -> None:
        self.name = name
        self.country = country
        self.pot = pot
        self.counter = 0
    
        self.opponents = dict()
        self.opponents["Pot 1"] = list()
        self.opponents["Pot 2"] = list()
        self.opponents["Pot 3"] = list()
        self.opponents["Pot 4"] = list() #rjecnik od 4 liste za svaki pot po jedna

        self.stadiums = dict()
        self.stadiums["Pot 1"] = [("Home", 0), ("Away", 0)]
        self.stadiums["Pot 2"] = [("Home", 0), ("Away", 0)]
        self.stadiums["Pot 3"] = [("Home", 0), ("Away", 0)]
        self.stadiums["Pot 4"] = [("Home", 0), ("Away", 0)]

        self.done = False
        return

    def drawOpponent(self, club: Club) -> None: # if 0 then failed, if 1 then OK

        if (club.country == self.country) or self.opponents[club.pot].__len__() == 2:
            return
        
        if self.stadiums[club.pot][0][1] == 0 and club.stadiums[self.pot][1][1] == 0: #vanjski klub "club" ide u goste trenutnom klubu
            club.opponents[self.pot].append(self)
            # club.stadiums[self.pot][1][1] = 1
            club.stadiums[self.pot][1] = ("Away", 1)
            self.opponents[club.pot].append(club)
            # self.stadiums[club.pot][0][1] = 1
            self.stadiums[club.pot][0] = ("Home", 1)
            self.counter += 1
        elif self.stadiums[club.pot][1][1] == 0 and club.stadiums[self.pot][0][1] == 0: #vanjski klub "club" igra doma protiv trenutnog kluba
            club.opponents[self.pot].append(self)
            # club.stadiums[self.pot][0][1] = 1
            club.stadiums[self.pot][0] = ("Home", 1)
            self.opponents[club.pot].append(club)
            # self.stadiums[club.pot][1][1] = 1
            self.stadiums[club.pot][1] = ("Away", 1)
            self.counter += 1
        else:
            return

        if self.counter == 8:
            self.done = True
        
        return

# Club("saw", "efhshe", "1").drawOpponent("asa")

a = list()
a.__len__()
```

```python
from datetime import datetime
import pandas as pd
import sys

class Club:
    def __init__(self, name: str, country: str, pot: str) -> None:
        self.name = name
        self.country = country
        self.pot = pot
        self.counter = 0

        self.opponents = dict()
        self.opponents["Pot 1"] = list()
        self.opponents["Pot 2"] = list()
        self.opponents["Pot 3"] = list()
        self.opponents["Pot 4"] = list() #rjecnik od 4 liste za svaki pot po jedna

        self.stadiums = dict()
        self.stadiums["Pot 1"] = [("Home", 0), ("Away", 0)]
        self.stadiums["Pot 2"] = [("Home", 0), ("Away", 0)]
        self.stadiums["Pot 3"] = [("Home", 0), ("Away", 0)]
        self.stadiums["Pot 4"] = [("Home", 0), ("Away", 0)]

        self.done = False
        return

'------------------------------------------------------------------------------------'


class Match:
    def __init__(self, home: Club, away: Club, date: datetime = 0) -> None:
        self.home = home
        self.away = away
        self.date = date
        return
    
    def show(self) -> None:
        output = "\n" + self.home.name + " will host match againts " + self.away.name + ", on " + str(self.date)
        sys.stdout.write(output)

'------------------------------------------------------------------------------------'

class Draw:
    def __init__(self, path: str = "C:\\Users\\kvrankic\\Desktop\\UefaDraw\\qualified_clubs.xlsx") -> None:

        df = pd.read_excel(path).sort_values('Status')
        df.drop(df[df['Status'] == 'Unqualified'].index, inplace=True)
        df.sort_values('Pot', inplace=True)
        df.index += 1
        self.competitors = df
        
        df = self.competitors.sort_values('Name')
        df['Points'] = 0
        df['Won'] = 0
        df['Draw'] = 0
        df['Lost'] = 0
        df['Goal Difference'] = 0
        df.sort_values(['Points', 'Goal Difference'], ascending=[False, False])
        df = df.reset_index(drop=True)
        df.index += 1
        self.table = df

        self.clubs = self.competitors['Name'].tolist()
        self.matchbook = dict() # rjecnik di je kljuc ime kluba a value lista Match-eva
        for club in self.clubs:
            self.matchbook[club] = [0] * 8
        
        countryList = self.competitors['Country'].tolist()
        potList = self.competitors['Pot'].tolist()
        clubList = list()
        for i, club in enumerate(self.clubs):
            clubList.append(Club(name=club, country=countryList[i], pot=potList[i]))
        self.clubs = clubList

        # sad je zajeb, radis li sa pandas poljima ili klasama klub
        potentialOpponents = dict()
        for club in self.clubs:
            counter = 0
            for iterClub in self.clubs:
                if (club.country == iterClub.country):
                    continue
                counter += 1
            potentialOpponents[club] = counter
        
        self.potentialOpponents = potentialOpponents
        
        return
    
    
    def exportTable(self, name: str = 'uefa_table', path: str = 'C:\\Users\\kvrankic\\Desktop\\UefaDraw', csv: bool = False) -> None:

        if csv:
            self.table.to_csv(path+name+'.csv')
        else:
            self.table.to_excel(path+name+'.xlsx')

        return
    
    def makeMatch(self, club1, club2) -> None: #1 -> 0,1 ; #2 -> 2,3; #3 -> 4,5; #4 -> 6,7
        
        if self.matchbook[club1][(club2.pot * 2) - 2] == 0: # club1 home, club2 away
            match = Match(club1, club2)
            self.matchbook[club1][(club2.pot * 2) - 2] = match
            self.matchbook[club2][(club1.pot * 2) - 1] = match
        elif self.matchbook[club1][(club2.pot * 2) - 1] == 0: # club1 away, club2 home
            match = Match(club2, club1)
            self.matchbook[club1][(club2.pot * 2) - 1] = match
            self.matchbook[club2][(club1.pot * 2) - 2] = match

        return

    def showMatches(self) -> None:

        for club in self.matchbook:
            for i in range(len(self.matchbook[club])):
                self.matchbook[club][i].show() #not sure jel se ovako metoda koristi
        return
    
    

```
